## Dados de entrada

```python
W = 574000
T = 450
dr = 41.5
d = 50
p = 8
b = 4
H = 48

print(f'Carga axial no prafuso: {W} N')
print(f'Momento de torção: {T} Nm')
print(f'Diâmetro menor: {dr} mm')
print(f'Diâmetro maior: {d} mm')
print(f'Passo: {p} mm')
print(f'Largura do filete na raiz: {b} mm')
print(f'Comprimento da parte da porca que está sob carga: {H} mm')
```

## Número de voltas no filete (n)

```python
def n(H,p):
  n = H/p
  return n
```

```python
n(H,p)
```

## Altura do filete (h)

```python
def h(d,dr):
  h = (d-dr)/2
  return h
```

## Raio médio (rm)

```python
def rm(d,dr):
  rm = (d+dr)/4
  return rm
```

## Código principal

```python
tensao_cis_rais = W / (2 * 3.14 * n(H,p) * rm(d,dr) * b)
tensao_compr = W / (2 * 3.14 * n(H,p) * rm(d,dr) * h(d,dr))
tensao_cis_cilindro = ((16 * T) / (3.14*((dr*10**-3)**3)))*10**-6
tensao_normal_cilindro = ((4*W) / (3.14*((dr*10**-3)**2)))*10**-6

print(f'Tensão de cisalhamento na raiz do parafuso: {tensao_cis_rais:.2f} MPa')
print(f'Tensão de compressão entre os filetes da porca e do parafuso: {tensao_compr:.2f} MPa')
print(f'Tensão de cisalhamento no cilindro da base e do parafuso: {tensao_cis_cilindro:.2f} MPa')
print(f'Tensão normal no cilindro de base: {tensao_normal_cilindro:.2f} Mpa')
```
