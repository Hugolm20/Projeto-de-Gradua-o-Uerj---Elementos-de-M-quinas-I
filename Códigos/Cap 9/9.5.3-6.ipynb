```python
D = 102 /1000
L = 152/1000
C = 1.02 * (10**-4)
W = 5340
rpm = 500
T = 63
Ta = 32
#olhar tabela L/D para achar valor de k
k = 2.03125* (10**-3)
ventilacao = 'sim'
```

```python
# encontrar o valor de K
if ventilacao == 'sim':
  K = 31
else: 
  K = 55
```

```python
delta_t = T - Ta
print(delta_t)
```

```python
# quantidade de calor dissipada
Qd = (113.015/K) * ((delta_t + 18.3)**2) * L * D 
print(Qd)
```

```python
#coeficiente de atrito
#fazendo Qg = Qd 
f = (60 * Qd) / (W * 3.14 * D * (rpm))
print(f)
```

```python
#pressão sobre o mancal 
x = L * D
p = W / x
print(p)
```

```python
#viscosidade dinâmica
u = p * ((f - k) / (0.334 * (rpm) * (D/C))  )
print(u)
```

```python
#temperatura do óleo lubrificante
To = Ta + (2*delta_t)
```

```python
#número de sommerfield
S = ((u* (rpm/60))/p)*(D/C)**2
```

```python
#massa específica do lubrificante
ro = 900 - (0.365 * (To - 60))
```

```python
print(f'μ = {u} Pa.s')
print(f'To = {To} °C')
print(f'S = {S}')
print(f'ρ = {ro}')
```
