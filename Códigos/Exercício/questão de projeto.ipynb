```python
kp = 1 + (q * (kt - 1))
print(kp)
#ka
if aco == 'forjado':
    a = 272
    b = -0.995
    ka = a * (tensao_ruptura ** b)
    print(f'ka = {ka:.2f}')
elif aco == 'laminado_quente':
    a = 57.7
    b = -0.718
    ka = a * (tensao_ruptura ** b)
    print(f'ka = {ka:.2f}')
elif aco == 'usinado_ou_laminado_frio':
    a = 4.51
    b = -0.265
    ka = a * (tensao_ruptura ** b)
    print(f'ka = {ka:.2f}')
elif aco == 'retificado':
    a = 1.58
    b = -0.085
    ka = a * (tensao_ruptura ** b)
    print(f'ka = {ka:.2f}')
elif aco == 'espelhado':
    a = 1
    b = 0
    ka = a * (tensao_ruptura ** b)
    print(f'ka = {ka:.2f}')


#kb
if d == '':
    kb = 0.85
    print(f'kb = {kb:.2f}')
elif 2.79*(10**-3) <=  d <= 51*(10**-3):
    kb = (d / 7.62)**-0.107
    print(f'kb = {kb:.2f}')
elif 51*(10**-3) < d <= 254*(10**-3):
    kb = 1.51 * (d**-0.157)
    print(f'kb = {kb:.2f}')

#kc 
if conf == 50:
    kc = 1
    print(f'kc = {kc:.2f}')
elif conf == 90:
    kc = 0.897
    print(f'kc = {kc:.2f}')
elif conf == 95:
    kc = 0.868
    print(f'kc = {kc:.2f}')
elif conf == 98:
    kc = 0.836
    print(f'kc = {kc:.2f}')
elif conf == 99:
    kc = 0,814
    print(f'kc = {kc:.2f}')
elif conf == 99.99:
    kc = 0.702
    print(f'kc = {kc:.2f}')
elif conf == 99.9999:
    kc = 0.620
    print(f'kc = {kc:.2f}')
elif conf == 99.99999:
    kc = 0.584
    print(f'kc = {kc:.2f}')

#kd
kd = 0.9877 + (0.6507 * (10**-3) * temp) - (0.3414 * (10**-5) * (temp**2)) + (0.5621 * (10**-8) * (temp**3)) - (6.246 * (10**-12) * (temp**4))
print(f'kd = {kd:.2f}')

#ke 
if carga == 'flexao':
    ke = 1
    print(f'ke = {ke:.2f}')
elif carga == 'axial':
    ke = 0.85
    print(f'ke = {ke:.2f}')
elif carga == 'torcao':
    ke = 0.577
    print(f'ke = {ke:.2f}')

#kf 
kf = 1
print(f'kf = {kf:.2f}')


lim_normal_res = ka * kb * kc * kd * 0.85 * kf * (0.5 * tensao_ruptura)
lim_cisalhante_res = ka * kb * kc * kd * 0.577 * kf * (0.5 * tensao_ruptura)

print(lim_normal_res)
print(lim_cisalhante_res)

i = 0
a = []

while i < len(matriz_posicao):
    prod = np.cross(matriz_posicao[i], matriz_forca[i])
    prod = a.append(prod)
    prod = sum(a)
    i = i + 1

if matriz_momento == '':
    print(prod)
else:
    matriz_momento = np.sum(matriz_momento, axis=0)
    prod = prod + matriz_momento
    print(prod)

matriz_forca = np.sum(matriz_forca, axis=0)
print(f'F = {matriz_forca[0]:.2f} i + {matriz_forca[1]:.2f} j + {matriz_forca[2]:.2f} k')
print(f'M = {prod[0]:.2f} i + {prod[1]:.2f} j + {prod[2]:.2f} k')

t = prod[2]
m = math.sqrt(((prod[0])**2) + ((prod[1])**2))

print(f'Momento torçor [T]: {t}')
print(f'Momento fletor [M]: {m}')

tensao_cisalhante_esc = 0.577 * lim_esc
print(tensao_cisalhante_esc)
```

```python
#aplicar método de newton raphson na equação do diâmetro
def newton(f, flin, d0, epsilon, maxIter=20):
  if abs(f(d0)) <= epsilon:
    return (d0)

  print("k\t d0\t\t f(d0)") #cabeçalho da tabela que retornará com os valores embaixo. \t é o espaçamento que dou para ficar visual.
  
  k=1
  while k <= maxIter:
    d1 = d0 - f(d0)/flin(d0)

    # % - indica o marcador de posição. usada para indicar a composição da string com o conteúdo da variável
    # %d - indica que o conteúdo vai ser inteiro
    # %e - indica que o conteúdo vai ser um exponencial  
    print("%d\t%e\t%e" % (k, d1,f(d1)))

    if abs(f(d1)) <= epsilon:
      return d1      
    
    d0 = d1
    k = k + 1
  
  print('ERRO: Número máximo de iterações atingido')
  
  return d1
    
if __name__ == "__main__": 
  def f(d):
    return math.sqrt(((tensao_normal/(d**3))**2) + (3*(tensao_cisalhante/(d**3))**2))

  def flin(d):
    return (-1055.16247) / (d**4)

```
