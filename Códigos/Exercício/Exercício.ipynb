```python
import numpy as np 
import math
```

# Dados de entrada

```python
matriz_posicao = [[0.28, 0, 0.3]]
matriz_forca  = [[0, 16, 12]]
matriz_momento = ''
d = 0.08
tensao_ruptura = 1000
lim_esc = 800
aco = 'forjado'
carga = 'axial'
conf = 98
q = 1
kt = 1
n = 2
temp = 80
print(f'Matriz posição: {matriz_posicao}')
print(f'Matriz força: {matriz_forca}')
print(f'Matriz momento: {matriz_momento}')
print(f'Diâmetro: {d}m')
print(f'Tensão de ruptura: {tensao_ruptura} MPa')
print(f'Tipo de aço: {aco}')
print(f'Temperatura: {temp}°C')
print(f'Tipo de carga: {carga}')
print(f'Confiabilidade %: {conf}')
print(f'Fator teórico de tensão: {kt}')
```

# Resolução do exercício

```python
if d == '':
    kp = 1 + (q * (kt - 1))
    print(f'kp: {kp}')
    #ka
    if aco == 'forjado':
        a = 272
        b = -0.995
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'laminado_quente':
        a = 57.7
        b = -0.718
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'usinado_ou_laminado_frio':
        a = 4.51
        b = -0.265
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'retificado':
        a = 1.58
        b = -0.085
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'espelhado':
        a = 1
        b = 0
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')


    #kb
    if d == '':
        kb = 0.85
        print(f'kb = {kb:.2f}')
    elif 2.79*(10**-3) <=  d <= 51*(10**-3):
        kb = (d / 7.62)**-0.107
        print(f'kb = {kb:.2f}')
    elif 51*(10**-3) < d <= 254*(10**-3):
        kb = 1.51 * (d**-0.157)
        print(f'kb = {kb:.2f}')

    #kc 
    if conf == 50:
        kc = 1
        print(f'kc = {kc:.2f}')
    elif conf == 90:
        kc = 0.897
        print(f'kc = {kc:.2f}')
    elif conf == 95:
        kc = 0.868
        print(f'kc = {kc:.2f}')
    elif conf == 98:
        kc = 0.836
        print(f'kc = {kc:.2f}')
    elif conf == 99:
        kc = 0,814
        print(f'kc = {kc:.2f}')
    elif conf == 99.99:
        kc = 0.702
        print(f'kc = {kc:.2f}')
    elif conf == 99.9999:
        kc = 0.620
        print(f'kc = {kc:.2f}')
    elif conf == 99.99999:
        kc = 0.584
        print(f'kc = {kc:.2f}')

    #kd
    kd = 0.9877 + (0.6507 * (10**-3) * temp) - (0.3414 * (10**-5) * (temp**2)) + (0.5621 * (10**-8) * (temp**3)) - (6.246 * (10**-12) * (temp**4))
    print(f'kd = {kd:.2f}')

    #ke 
    if carga == 'flexao':
        ke = 1
        print(f'ke = {ke:.2f}')
    elif carga == 'axial':
        ke = 0.85
        print(f'ke = {ke:.2f}')
    elif carga == 'torcao':
        ke = 0.577
        print(f'ke = {ke:.2f}')

    #kf 
    kf = 1
    print(f'kf = {kf:.2f}')


    lim_normal_res = ka * kb * kc * kd * 0.85 * kf * (0.5 * tensao_ruptura)
    lim_cisalhante_res = ka * kb * kc * kd * 0.577 * kf * (0.5 * tensao_ruptura)

    print(f'Limite normal de resistência: {lim_normal_res}')
    print(f'Limite cisalhante de resistência: {lim_cisalhante_res}')

    print('-'*50)

    i = 0
    a = []

    while i < len(matriz_posicao):
        prod = np.cross(matriz_posicao[i], matriz_forca[i])
        prod = a.append(prod)
        prod = sum(a)
        i = i + 1

    if matriz_momento == '':
        print(prod)
    else:
        matriz_momento = np.sum(matriz_momento, axis=0)
        prod = prod + matriz_momento
        print(prod)

    matriz_forca = np.sum(matriz_forca, axis=0)
    print(f'F = {matriz_forca[0]:.2f} i + {matriz_forca[1]:.2f} j + {matriz_forca[2]:.2f} k')
    print(f'M = {prod[0]:.2f} i + {prod[1]:.2f} j + {prod[2]:.2f} k')

    t = prod[2]
    m = math.sqrt(((prod[0])**2) + ((prod[1])**2))

    print(f'Momento torçor [T]: {t}')
    print(f'Momento fletor [M]: {m}')

    print('-'*50)

    tensao_cisalhante_esc = 0.577 * lim_esc
    print(f'Tensão cisalhante de escoamento: {tensao_cisalhante_esc}')

    tensao_normal_max = kp * ((32 * m) / 3.14)
    tensao_normal_min = 0
    tensao_cisalhante_max = kp * ((16 * t) / 3.14)
    tensao_cisalhante_min = 0

    tensao_normal_var = (tensao_normal_max - tensao_normal_min) / 2
    tensao_normal_media = (tensao_normal_max + tensao_normal_min) / 2

    tensao_cisalhante_media = (tensao_cisalhante_max + tensao_cisalhante_min) / 2
    tensao_cisalhante_var = (tensao_cisalhante_max - tensao_cisalhante_min) / 2

    tensao_normal = tensao_normal_media + (tensao_normal_var * (lim_esc/lim_normal_res))
    tensao_cisalhante = tensao_cisalhante_media + (tensao_cisalhante_var * (tensao_cisalhante_esc/lim_cisalhante_res))

    print(f'Tensão normal máxima: {tensao_normal_max}')
    print(f'Tensão cisalhante máxima: {tensao_cisalhante_max}')
    print(f'Tensão normal variável: {tensao_normal_var}')
    print(f'Tensão normal média: {tensao_normal_media}')
    print(f'Tensão cisalhante média: {tensao_cisalhante_media}')
    print(f'Tensão cisalhante variável: {tensao_cisalhante_var}')
    print(f'Tensão normal: {tensao_normal}')
    print(f'Tensão cisalhante: {tensao_cisalhante}')

    d_0 = ((math.sqrt((tensao_normal**2)+(3*(tensao_cisalhante**2)))*n)/lim_esc)**1/3

    print('-'*50)

    print(f'Diâmetro inicial: {d_0}')

    def newton(f, flin, d0, epsilon, maxIter=20):
      if abs(f(d0)) <= epsilon:
        return (d0)

      print("k\t d0\t\t f(d0)") #cabeçalho da tabela que retornará com os valores embaixo. \t é o espaçamento que dou para ficar visual.
      
      k=1
      while k <= maxIter:
        d1 = d0 - f(d0)/flin(d0)

        # % - indica o marcador de posição. usada para indicar a composição da string com o conteúdo da variável
        # %d - indica que o conteúdo vai ser inteiro
        # %e - indica que o conteúdo vai ser um exponencial  
        print("%d\t%e\t%e" % (k, d1,f(d1)))

        if abs(f(d1)) <= epsilon:
          return d1      
        
        d0 = d1
        k = k + 1
      
      print('ERRO: Número máximo de iterações atingido')
      
      return d1
    
    if __name__ == "__main__": 
      def f(d):
        return math.sqrt(((tensao_normal/(d**3))**2) + (3*(tensao_cisalhante/(d**3))**2))

      def flin(d):
        return (-1055.16247) / (d**4)

    raiz = newton(f,flin,d_0,10000)
    print(f'Diâmetro: {raiz}')


else:
    kp = 1 + (q * (kt - 1))
    print(f'kp: {kp}')

    #ka
    if aco == 'forjado':
        a = 272
        b = -0.995
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'laminado_quente':
        a = 57.7
        b = -0.718
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'usinado_ou_laminado_frio':
        a = 4.51
        b = -0.265
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'retificado':
        a = 1.58
        b = -0.085
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')
    elif aco == 'espelhado':
        a = 1
        b = 0
        ka = a * (tensao_ruptura ** b)
        print(f'ka = {ka:.2f}')


    #kb
    if d == '':
        kb = 0.85
        print(f'kb = {kb:.2f}')
    elif 2.79*(10**-3) <=  d <= 51*(10**-3):
        kb = (d / 7.62)**-0.107
        print(f'kb = {kb:.2f}')
    elif 51*(10**-3) < d <= 254*(10**-3):
        kb = 1.51 * (d**-0.157)
        print(f'kb = {kb:.2f}')

    #kc 
    if conf == 50:
        kc = 1
        print(f'kc = {kc:.2f}')
    elif conf == 90:
        kc = 0.897
        print(f'kc = {kc:.2f}')
    elif conf == 95:
        kc = 0.868
        print(f'kc = {kc:.2f}')
    elif conf == 98:
        kc = 0.836
        print(f'kc = {kc:.2f}')
    elif conf == 99:
        kc = 0,814
        print(f'kc = {kc:.2f}')
    elif conf == 99.99:
        kc = 0.702
        print(f'kc = {kc:.2f}')
    elif conf == 99.9999:
        kc = 0.620
        print(f'kc = {kc:.2f}')
    elif conf == 99.99999:
        kc = 0.584
        print(f'kc = {kc:.2f}')

    #kd
    kd = 0.9877 + (0.6507 * (10**-3) * temp) - (0.3414 * (10**-5) * (temp**2)) + (0.5621 * (10**-8) * (temp**3)) - (6.246 * (10**-12) * (temp**4))
    print(f'kd = {kd:.2f}')

    #ke 
    if carga == 'flexao':
        ke = 1
        print(f'ke = {ke:.2f}')
    elif carga == 'axial':
        ke = 0.85
        print(f'ke = {ke:.2f}')
    elif carga == 'torcao':
        ke = 0.577
        print(f'ke = {ke:.2f}')

    #kf 
    kf = 1
    print(f'kf = {kf:.2f}')

    lim_normal_res = ka * kb * kc * kd * 0.85 * kf * (0.5 * tensao_ruptura)
    lim_cisalhante_res = ka * kb * kc * kd * 0.577 * kf * (0.5 * tensao_ruptura)

    print(f'Limite normal de resistência: {lim_normal_res}')
    print(f'Limite cisalhante de resistência: {lim_cisalhante_res}')

    print('-'*50)

    i = 0
    a = []

    while i < len(matriz_posicao):
        prod = np.cross(matriz_posicao[i], matriz_forca[i])
        prod = a.append(prod)
        prod = sum(a)
        i = i + 1

    if matriz_momento == '':
        print(prod)
    else:
        matriz_momento = np.sum(matriz_momento, axis=0)
        prod = prod + matriz_momento
        print(prod)

    matriz_forca = np.sum(matriz_forca, axis=0)
    print(f'F = {matriz_forca[0]:.2f} i + {matriz_forca[1]:.2f} j + {matriz_forca[2]:.2f} k')
    print(f'M = {prod[0]:.2f} i + {prod[1]:.2f} j + {prod[2]:.2f} k')

    t = prod[2]
    m = math.sqrt(((prod[0])**2) + ((prod[1])**2))

    print(f'Momento torçor [T]: {t}')
    print(f'Momento fletor [M]: {m}')

    print('-'*50)

    tensao_cisalhante_esc = 0.577 * lim_esc
    print(f'Tensão cisalhante de escoamento: {tensao_cisalhante_esc}')
      
    tensao_normal_max = kp * ((32 * m) /(d * 3.14))
    tensao_normal_min = 0
    tensao_cisalhante_max = kp * ((16 * t) / (d * 3.14))
    tensao_cisalhante_min = 0

    print(f'Tensão normal mínima: {tensao_normal_min:.2f} MPa')    
    print(f'Tensão normal máxima: {tensao_normal_max:.2f} MPa')    
    print(f'Tensão cisalhante máxima: {tensao_cisalhante_max:.2f} MPa')    
    print(f'Tensão cisalhante mínima: {tensao_cisalhante_min:.2f} MPa')

    print('-'*50)

    tensao_normal_media = (tensao_normal_max + tensao_normal_min) / 2 
    tensao_normal_var = (tensao_normal_max - tensao_normal_min) / 2
    
    tensao_cisalhante_media = (tensao_cisalhante_max + tensao_cisalhante_min) / 2
    tensao_cisalhante_var = (tensao_cisalhante_max - tensao_cisalhante_min) / 2

    print(f'Tensão normal variável: {tensao_normal_var}')
    print(f'Tensão normal média: {tensao_normal_media}')
    print(f'Tensão cisalhante média: {tensao_cisalhante_media}')
    print(f'Tensão cisalhante variável: {tensao_cisalhante_var}')

    print('-'*50)

    #encontrar as tensões equivalentes utilizando um critério de fadiga arbitrariamente escolhido

    tensao_normal_sorderberg = tensao_normal_media + (tensao_normal_var * (lim_esc / lim_normal_res))
    tensao_cisalhante_sorderberg = tensao_cisalhante_media + (tensao_cisalhante_var * (tensao_cisalhante_esc / lim_cisalhante_res))
    fat_seg_sorderberg = lim_esc / ((tensao_normal_sorderberg**2) + (3 * (tensao_cisalhante_sorderberg**2)))**(1/2)

    print(f'Tensão normal sorderberg: {tensao_normal_sorderberg}')
    print(f'Tensão cisalhante sorderberg: {tensao_cisalhante_sorderberg}')
    print(f'N sorderberg: {fat_seg_sorderberg}')

    print('-'*50)

    tensao_normal_mod_god = lim_esc * ((tensao_normal_var/lim_normal_res) + (tensao_normal_media/tensao_ruptura))
    tensao_cisalhante_mod_god = tensao_cisalhante_esc * ((tensao_cisalhante_var/lim_cisalhante_res) + (tensao_cisalhante_media/(tensao_ruptura*0.8)))
    fat_seg_mod_god = lim_esc / (((tensao_normal_mod_god**2) + 3 * (tensao_cisalhante_mod_god**2))**(1/2))

    print(f'Tensão normal modificado de goodman: {tensao_normal_mod_god}')
    print(f'Tensão cisalhante modificado de goodman: {tensao_cisalhante_mod_god}')
    print(f'N modificado de goodman: {fat_seg_mod_god}')

    print('-'*50)

    a1 = 1
    b1 = (lim_esc * tensao_normal_var) / lim_normal_res
    c1 = -((lim_esc * tensao_normal_media) / tensao_ruptura)**2
    
    if(a1==0):
        print('a=0, não é equação do segundo grau')
    else:
        delta1 = b1*b1 - (4*a1*c1)

        if delta1<0:
            print('Delta menor que 0. Raízes imaginárias')
        elif delta1==0:
            tensao_normal_gerber = -b1 / (2*a1)
            print('Delta=0 , raiz = ',tensao_normal_gerber)
        else:
            tensao_normal_gerber1 = (-b1 + math.sqrt(delta1) ) / (2*a1)
            tensao_normal_gerber2 = (-b1 - math.sqrt(delta1) ) / (2*a1)
            print(f'Tensão normal de gerber: {tensao_normal_gerber1}' )

    a2 = 1
    b2 = (tensao_cisalhante_esc * tensao_cisalhante_var) / lim_cisalhante_res
    c2 = -((tensao_cisalhante_esc * tensao_cisalhante_media) / (tensao_ruptura * 0.8))**2 

    if(a2==0):
        print('a=0, não é equação do segundo grau')
    else:
        delta2 = b2*b2 - (4*a2*c2)

        if delta2<0:
            print('Delta menor que 0. Raízes imaginárias')
        elif delta2==0:
            tensao_cisalhante_gerber = -b2 / (2*a2)
            print('Delta=0 , raiz = ',tensao_cisalhante_gerber)
        else:
            tensao_cisalhante_gerber1 = (-b2 + math.sqrt(delta2) ) / (2*a2)
            tensao_cisalhante_gerber2 = (-b2 - math.sqrt(delta2) ) / (2*a2)
            print(f'Tensão cisalhante de gerber: {tensao_cisalhante_gerber1}' )

    fat_seg_gerber = lim_esc / (((tensao_normal_gerber1 **2) + 3 * (tensao_cisalhante_gerber1 **2))**(1/2))
    print(f'N gerber: {fat_seg_gerber}')

    print('-'*50)

    tensao_normal_asme = math.sqrt((((lim_esc * tensao_normal_var)/lim_normal_res)**2) + ((tensao_normal_media)**2))
    tensao_cisalhante_asme = math.sqrt((((tensao_cisalhante_esc * tensao_cisalhante_var)/lim_cisalhante_res)**2) + ((tensao_cisalhante_media)**2))
    fat_seg_asme = lim_esc/math.sqrt((tensao_normal_asme**2) + 3*(tensao_cisalhante_asme**2))

    print(f'Tensão normal ASME: {tensao_normal_asme}')
    print(f'Tensão cisalhante ASME: {tensao_cisalhante_asme}')
    print(f'N ASME: {fat_seg_asme}')

    print('-'*50)

    tensao_normal_escamp = tensao_normal_var + tensao_normal_media
    tensao_cisalhante_escamp = tensao_cisalhante_var + tensao_cisalhante_media
    fat_seg_escamp = lim_esc/math.sqrt((tensao_normal_asme**2) + 3*(tensao_cisalhante_escamp**2))
    
    print(f'Tensão normal escoamento ampliado: {tensao_normal_escamp}')
    print(f'Tensão cisalhante escoamento ampliado: {tensao_cisalhante_escamp}')
    print(f'N escoamento ampliado {fat_seg_escamp}')
```
