```python
import math

lim_rup = 1000
lim_esc = 800
d = 0.04911
n = 2
q = 0.68
kt = 1.9

f_min = -180000
f_max = 180000
ka_n = 0.72
kb_n = 1
kc_n = 1
kd_n = 1
ke_n = 0.85
kf_n = 1

t_min = 0
t_max = 0
ka_c = 0.72
kb_c = 0.759
kc_c = 1
kd_c = 1
ke_c = 0.577
kf_c = 1


tensao_normal_max = (4*f_max)/(3.14 * (d**2)) * 10**-6
tensao_normal_min = (4*f_min)/(3.14 * (d**2)) * 10**-6

tensao_normal_media = (tensao_normal_max + tensao_normal_min) / 2 
tensao_normal_var = (tensao_normal_max - tensao_normal_min) / 2

fat_seg_1 = lim_esc / tensao_normal_max

lim_normal_res = ka_n * kb_n * kc_n * kd_n * ke_n * kf_n * (0.5 * lim_rup)

fat_seg_linha = (lim_normal_res * lim_esc) / ((tensao_normal_media * lim_normal_res) + (tensao_normal_var * lim_esc))

tensao_normal = tensao_normal_media + (tensao_normal_var * (lim_esc / lim_normal_res))

lim_cis_res = ka_c * kb_c * kc_c * kd_c * ke_c * kf_c * (0.5 * lim_rup)



tensao_cis_esc = 0.577 * lim_esc

tensao_cis_max = (16*t_max) / (3.14 * (d**3)) * 10**-6
tensao_cis_min = (16*t_min) / (3.14 * (d**3)) * 10**-6

tensao_cis_media = (tensao_cis_max + tensao_cis_min) / 2
tensao_cis_var = (tensao_cis_max - tensao_cis_min) / 2

tensao_cis = tensao_cis_media + (tensao_cis_var * (tensao_cis_esc / lim_cis_res))

fat_seg_2 = lim_esc / ((tensao_normal**2) + (3 * (tensao_cis**2)))**(1/2)


tensao_normal_mod_god = lim_esc * ((tensao_normal_var/lim_normal_res) + (tensao_normal_media/lim_rup))
tensao_cis_mod_god = tensao_cis_esc * ((tensao_cis_var/lim_cis_res) + (tensao_cis_media/lim_rup))
n_mod_god = lim_esc / (((tensao_normal_mod_god**2) + 3 * (tensao_cis_mod_god**2))**(1/2))

tensao_normal_asme = math.sqrt((((lim_esc * tensao_normal_var)/lim_normal_res)**2) + ((tensao_normal_media)**2))
tensao_cis_asme = math.sqrt((((tensao_cis_esc * tensao_cis_var)/lim_cis_res)**2) + ((tensao_cis_media)**2))
n_asme = lim_esc/math.sqrt((tensao_normal_asme**2) + 3*(tensao_cis_asme**2))

tensao_normal_escamp = tensao_normal_var + tensao_normal_media
tensao_cis_escamp = tensao_cis_var + tensao_cis_media
n_escamp = lim_esc/math.sqrt((tensao_normal_asme**2) + 3*(tensao_cis_asme**2))

print(f'Valor máximo da tensão normal : {tensao_normal_max:.2f}')
print(f'Valor mínimo da tensão normal : {tensao_normal_min:.2f}')
print(f'Tensão normal média : {tensao_normal_media:.2f}')
print(f'Tensão normal variável : {tensao_normal_var:.2f}')
print(f"N': {fat_seg_1:.2f}")
print(f"N' = {fat_seg_linha:.2f}")
print(f'Tensão normal: {tensao_normal:.2f}')
print(f'Valor máximo da tensão cisalhamento : {tensao_cis_max:.2f}')
print(f'Valor mínimo da tensão cisalhamento : {tensao_cis_min:.2f}')
print(f'Tensão cisalhamento média : {tensao_cis_media:.2f}')
print(f'Tensão cisalhamento variável : {tensao_cis_var:.2f}')
print(f'Tensão cisalhamento : {tensao_cis:.2f}')
print(f'N = {fat_seg_2:.2f}')
print('-' * 100)
print(f'N Modificado de Godman: {n_mod_god:.2f}')
print(f'N ASME: {n_asme:.2f}')
print(f'N Escoamento Ampliado: {n_escamp:.2f}')
```
