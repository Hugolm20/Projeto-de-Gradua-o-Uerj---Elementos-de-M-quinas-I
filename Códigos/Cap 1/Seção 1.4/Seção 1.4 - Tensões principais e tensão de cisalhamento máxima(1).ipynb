```python
try:     
    tensao_normal_1 = float(input('Insira o valor de σ1:'))
    tensao_normal_2 = float(input('Insira o valor de σ2:'))
    tensao_cisalhante_12 = float(input('Insira o valor de τ12'))
except ValueError as error:
    print('O valor deve ser um número')      
```

```python
# letra a
tan_2_teta_n = (2 * tensao_cisalhante_12) / (tensao_normal_1 - tensao_normal_2)
dois_teta_n = math.degrees(math.atan(tan_2_teta_n))
θn1 = dois_teta_n / 2
θn2 = θn1 + 90 

# letra b
tensao_normal_max = ((tensao_normal_1 + tensao_normal_2) / 2) + math.sqrt(((tensao_normal_1 - tensao_normal_2) / 2)**2 + tensao_cisalhante_12**2)
tensao_normal_min = ((tensao_normal_1 + tensao_normal_2) / 2) - math.sqrt(((tensao_normal_1 - tensao_normal_2) / 2)**2 + tensao_cisalhante_12**2)

# letra c 
tan_2_teta_t = -((tensao_normal_1 - tensao_normal_2) / (2 * tensao_cisalhante_12)) 
dois_teta_t = math.degrees(math.atan(tan_2_teta_t))
θt1 = dois_teta_t / 2
θt2 = θt1 + 90 

# letra d 
tensao_cisalhante_max = math.sqrt(((tensao_normal_1 - tensao_normal_2) / 2)**2 + tensao_cisalhante_12**2)

# letra e 
tensao_media = ((tensao_normal_1 + tensao_normal_2) / 2)
```

```python
print('a)')
print(f'θn1: {θn1:.2f}°')
print(f'θn2: {θn2:.2f}°')
print('--------------------------------')
print('b)')
print(f'σmáx: {tensao_normal_max:.2f} Pa')
print(f'σmáx: {tensao_normal_min:.2f} Pa')
print('--------------------------------')
print('c)')
print(f'θt1: {θt1:.2f}°')
print('--------------------------------')
print('d)')
print(f'τmáx: {tensao_cisalhante_max:.2f} Pa')
print('--------------------------------')
print('e)')
print(f'τmédia: {tensao_media:.2f} Pa')
```
